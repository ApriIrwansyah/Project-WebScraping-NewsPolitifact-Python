```python
import requests
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd
import sys
```

```python
# Extraction details of a web-page

# Specify number of pages upto which you want to scrape article

# =====================================
# url         = "https://www.politifact.com/factchecks/list/?page=1"
# resp        = requests.get(url)
# print(resp.content)
# =====================================

pageToGet   = 10
upperframe  = []

for page in range(1, pageToGet+1):
    print(f"processing page : {page}")
    url         = f"https://www.politifact.com/factchecks/list/?page={str(page)}"
    
    # an exception might be thrown, so the code should be in a try-except block
    try: # ini untuk mengeksekusi jika ada error
        page = requests.get(url)
        
    # Response is equivalent to an enter key in chrome
    except Exception as e:  # this decribes what to do if an exception is thrown
        error_type, error_obj, error_info = sys.exc_info() # get the exception information
        print(f'Error For Link : {url}') # print the link that caused the problem
        print(error_type, 'Line', error_info.tb_lineno) # print error info and line that threw the exeption
        continue
    
    # Delaying request time by 2 second
    sleep(2)
    
    # Parsing html using Beautifull Soup to pull data out of HTML & XML files
    soup        = BeautifulSoup(page.content, 'html.parser')
    
    frame       = [] # Membuat List
    
    # All links of news article listed on a page
    links       = soup.find_all('li', {'class': 'o-listicle__item'})    
    print(len(links))
    
    # Writing to a file
    filename    = "WebScraped.csv"
    f           = open(filename, 'w')
    headers     = "Statement, Link, Date, Source, Label \n"    
    f.write(headers)
    
    # Extraction Specific details of a web
    for j in links:
        statement       = j.find('div', {'class':'m-statement__quote'}).text
        link           = j.find('div', {'class':'m-statement__quote'}).find('a')['href'].strip()
        date           = j.find('div', {'class':'m-statement__body'}).find('footer').text[-14:-1].strip()
        source         = j.find('div', {'class':'m-statement__author'}).find('a').text.strip()
        label          = j.find('div', {'class':'m-statement__content'}).find('img', {'class':'c-image__original'}).get('src')
        
        frame.append((statement, link, date, source, label))
        f.write(statement.replace(",","^")+","+link.replace(",","^")+","+date.replace(",","^")+","
                +source.replace(",","^")+","+label.replace(",","^"))
        
    
# Do remember always to close your opened file after writing to it
f.close  

# Visualisasing Dataset Using Pandas Dataframe.
data = pd.DataFrame(upperframe, columns=['statement', 'link', 'date', 'source', 'label'])
data.head()
# data.shape
```
